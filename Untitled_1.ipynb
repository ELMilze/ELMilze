```python
X = data[['income', 'spending']]
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
labels = kmeans.labels_
data['Cluster'] = labels
segment_mapping = {
    0: 'Low Income- Low Spending Score',
    1: 'High Income- Low Spending Score',
    2: 'Low Income- High Spending Score',
    3: 'High Income- High Spending Score',
    4: 'Average Income- Average Spending Score'
}
data['Segment'] = data['Cluster'].map(segment_mapping)
print(data[['income', 'spending', 'Segment']])
```

```python
max_clusters = 10
cluster_range = range(1, max_clusters + 1)
inertia = []
```

```python
X = data[['income', 'spending']]

# Define a range of values for k (number of clusters)
k_values = range(1, 11)

# Initialize a list to store the WCSS for each k value
wcss = []

# Compute the within-cluster sum of squares (WCSS) for each k value
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Create the elbow graph
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')

# Display the plot
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)

# add segmentation at dataframe
data['segmentation'] = y_kmeans

data
```

```python
income = data['income']
spending_score = data['spending']
gender = data['gender']

# Create a color map for the genders
color_map = {0.0: 'blue', 1.0: 'red'}

# Create the scatter plot with different colors for each gender
plt.figure(figsize=(8, 6))
for g in set(gender):
    plt.scatter(income[gender == g], spending_score[gender == g], color=color_map[g], label=g)

plt.title('Income vs Spending Score (Gender)')
plt.xlabel('Income')
plt.ylabel('Spending Score')

# Add a legend to differentiate the genders
plt.legend()

# Display the plot
plt.show()
```

```python
segments_to_display = ['Low Income- Low Spending Score', 'High Income- Low Spending Score', 'Average Income- Average Spending Score' ,'Low Income- High Spending Score', 'High Income- High Spending Score']
filtered_data = data[data['Segment'].isin(segments_to_display)]

# Create the joint plot
sns.set(style='ticks')
g = sns.JointGrid(data=filtered_data, x='income', y='spending', hue='Segment')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.histplot, kde=True)

# Set the title and labels of the plot
g.fig.suptitle('Segment Joint Plot: Income vs Spending')
g.ax_joint.set_xlabel('Income')
g.ax_joint.set_ylabel('Spending')

# Display the plot
plt.show()
```
